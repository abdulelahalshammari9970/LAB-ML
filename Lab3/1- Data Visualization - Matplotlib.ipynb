Let's walk through a very simple example using two numpy arrays. You can also use lists, but most likely you'll be passing numpy arrays or pandas columns (which essentially also behave like arrays).

# Matplotlib Object Oriented Method
Now that we've seen the basics, let's break it all down with a more formal introduction of Matplotlib's Object Oriented API. This means we will instantiate figure objects and then call methods or attributes from that object.

## Introduction to the Object Oriented Method

The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it. 

To begin we create a figure instance. Then we can add axes to that figure:

```python
# Create Figure (empty canvas)  # creates a new empty figure to draw on
fig = plt.figure()  # initializes a figure object and stores it in fig

# Add set of axes to figure  # adds a plotting area (axes) inside the figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # positions axes using [left, bottom, width, height] (values from 0 to 1)

# Plot on that set of axes  # draws the graph inside the created axes
axes.plot(x, y, 'b')  # plots x vs y using a blue line

axes.set_xlabel('Set X Label')  # sets the x-axis label using axes method
axes.set_ylabel('Set y Label')  # sets the y-axis label using axes method
axes.set_title('Set Title')  # sets the title of the plot
```

Code is a little more complicated, but the advantage is that we now have full control of where the plot axes are placed, and we can easily add more than one axis to the figure:

```python
# Creates blank canvas  # creates an empty figure to place plots on
fig = plt.figure()  # initializes a figure object

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # adds the main axes covering most of the figure
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  # adds a smaller inset axes inside the main one

# Larger Figure Axes 1  # plotting on the main axes
axes1.plot(x, y, 'b')  # plots x vs y with a blue line
axes1.set_xlabel('X_label_axes1')  # sets x-axis label for main axes
axes1.set_ylabel('Y_label_axes1')  # sets y-axis label for main axes
axes1.set_title('Axes 1 Title')  # sets title for main axes

# Insert Figure Axes 2  # plotting on the inset axes
axes2.plot(y, x, 'r')  # plots y vs x with a red line
axes2.set_xlabel('X_label_axes2')  # sets x-axis label for inset axes
axes2.set_ylabel('Y_label_axes2')  # sets y-axis label for inset axes
axes2.set_title('Axes 2 Title')  # sets title for inset axes
```

## subplots()

The plt.subplots() object will act as a more automatic axis manager.

Basic use cases:

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes  # creates a figure and an axes object in one line
fig, axes = plt.subplots()  # returns (figure, axes) so you can customize the plot using axes methods

# Now use the axes object to add stuff to plot  # draws and labels the plot using the axes object
axes.plot(x, y, 'r')  # plots x vs y with a red line
axes.set_xlabel('x')  # sets the x-axis label
axes.set_ylabel('y')  # sets the y-axis label
axes.set_title('title')  # sets the plot title
```

Then you can specify the number of rows and columns when creating the subplots() object:

```python
# Empty canvas of 1 by 2 subplots  # creates a figure with 1 row and 2 side-by-side subplots
fig, axes = plt.subplots(nrows=1, ncols=2)  # returns the figure and an array of 2 axes objects (one for each subplot)
```

```python
# Axes is an array of axes to plot on
axes
```

We can iterate through this array:

```python
for ax in axes:  # loops through each subplot (each axes object) in the axes array
    ax.plot(x, y, 'b')  # plots x vs y on the current subplot using a blue line
    ax.set_xlabel('x')  # sets the x-axis label for the current subplot
    ax.set_ylabel('y')  # sets the y-axis label for the current subplot
    ax.set_title('title')  # sets the title for the current subplot

# Display the figure object  # shows the whole figure (all subplots) in the notebook output
fig  # displays the figure object (Jupyter renders it as an image)
```

A common issue with matplolib is overlapping subplots or figures. We ca use **fig.tight_layout()** or **plt.tight_layout()** method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:

```python
fig, axes = plt.subplots(nrows=1, ncols=2)  # creates a figure with 1 row and 2 subplots and returns fig and axes array

for ax in axes:  # loops through each subplot (axes object)
    ax.plot(x, y, 'g')  # plots x vs y using a green line on each subplot
    ax.set_xlabel('x')  # sets x-axis label for each subplot
    ax.set_ylabel('y')  # sets y-axis label for each subplot
    ax.set_title('title')  # sets title for each subplot

fig  # displays the figure in the notebook

plt.tight_layout()  # automatically adjusts spacing to prevent overlap between subplots
```

### Figure size, aspect ratio and DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

```python
fig = plt.figure(figsize=(8,4), dpi=100)  # creates a new figure with size 8x4 inches and resolution 100 DPI
```

The same arguments can also be passed to layout managers, such as the `subplots` function:

```python
fig, axes = plt.subplots(figsize=(4,2))  # creates a figure and a single axes with size 4x2 inches

axes.plot(x, y, 'r')  # plots x vs y using a red line
axes.set_xlabel('x')  # sets the x-axis label
axes.set_ylabel('y')  # sets the y-axis label
axes.set_title('title')  # sets the plot title
```

## Legends, labels and titles

Now that we have covered the basics of how to create a figure canvas and add axes instances to the canvas, let's look at how decorate a figure with titles, axis labels, and legends.

**Figure titles**

A title can be added to each axis instance in a figure. To set the title, use the `set_title` method in the axes instance:

```python
ax.set_title("title")  # sets the title for the current axes (plot)
```

**Axis labels**

Similarly, with the methods `set_xlabel` and `set_ylabel`, we can set the labels of the X and Y axes:

```python
ax.set_xlabel("x")  # sets the label for the x-axis
ax.set_ylabel("y")  # sets the label for the y-axis
```

### Legends

You can use the **label="label text"** keyword argument when plots or other objects are added to the figure, and then using the **legend** method without arguments to add the legend to the figure: 

```python
fig = plt.figure()  # creates a new empty figure (canvas)

ax = fig.add_axes([0,0,1,1])  # adds an axes that fills the whole figure area

ax.plot(x, x**2, label="x**2")  # plots x vs x^2 and assigns a label for the legend
ax.plot(x, x**3, label="x**3")  # plots x vs x^3 and assigns a label for the legend
ax.legend()  # displays the legend using the labels from the plotted lines
```

Notice how are legend overlaps some of the actual plot!

The **legend** function takes an optional keyword argument **loc** that can be used to specify where in the figure the legend is to be drawn. The allowed values of **loc** are numerical codes for the various places the legend can be drawn. See the [documentation page](https://matplotlib.org/stable/) for details. Some of the most common **loc** values are:

```python
# Lots of options.....  # examples of different legend positions using the loc parameter

ax.legend(loc=1)  # places the legend in the upper-right corner
ax.legend(loc=2)  # places the legend in the upper-left corner
ax.legend(loc=3)  # places the legend in the lower-left corner
ax.legend(loc=4)  # places the legend in the lower-right corner

# .. many more options are available  # there are other loc values/positions you can use

# Most common to choose  # the most commonly used option
ax.legend(loc=0)  # lets Matplotlib automatically choose the best legend location
fig  # displays the figure in the notebook output
```

## Setting colors, linewidths, linetypes

Matplotlib gives you *a lot* of options for customizing colors, linewidths, and linetypes. 

There is the basic MATLAB like syntax (which I would suggest you avoid using for more clairty sake:

### Colors with MatLab like syntax

With matplotlib, we can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where `'b'` means blue, `'g'` means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, 'b.-' means a blue line with dots:

```python
# MATLAB style line color and style  # demonstrates using MATLAB-like shorthand for colors and line styles
fig, ax = plt.subplots()  # creates a figure and a single axes object

ax.plot(x, x**2, 'b.-')  # plots x vs x^2 using blue color with dot markers and solid line
ax.plot(x, x**3, 'g--')  # plots x vs x^3 using green color with dashed line
```

### Colors with the color= parameter

We can also define colors by their names or RGB hex codes and optionally provide an alpha value using the `color` and `alpha` keyword arguments. Alpha indicates opacity.

```python
fig, ax = plt.subplots()  # creates a figure and a single axes object

ax.plot(x, x+1, color="blue", alpha=0.5)  # plots x vs x+1 with blue color and 50% transparency
ax.plot(x, x+2, color="#8B008B")  # plots x vs x+2 using a hex color code (dark magenta)
ax.plot(x, x+3, color="#FF8C00")  # plots x vs x+3 using a hex color code (dark orange)
```

### Line and marker styles

To change the line width, we can use the `linewidth` or `lw` keyword argument. The line style can be selected using the `linestyle` or `ls` keyword arguments:

```python
fig, ax = plt.subplots(figsize=(12,6))  # creates a figure with size 12x6 inches and one axes

ax.plot(x, x+1, color="red", linewidth=0.25)  # plots thin red line (very small width)
ax.plot(x, x+2, color="red", linewidth=0.50)  # plots slightly thicker red line
ax.plot(x, x+3, color="red", linewidth=1.00)  # plots normal thickness red line
ax.plot(x, x+4, color="red", linewidth=2.00)  # plots thick red line

# possible linestyle options ‘-’, ‘--’, ‘-.’, ‘:’, ‘steps’  # different line styles
ax.plot(x, x+5, color="green", lw=3, linestyle='-')  # solid green line with width 3
ax.plot(x, x+6, color="green", lw=3, ls='--')  # dashed green line
ax.plot(x, x+7, color="green", lw=3, ls=':')  # dotted green line

# custom dash  # defines a custom dash pattern
line, = ax.plot(x, x+8, color="black", lw=1.50)  # plots black line and stores it in variable
line.set_dashes([5, 10, 15, 10])  # custom pattern: dash length, gap, dash, gap...

# possible marker symbols marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...  # different marker styles
ax.plot(x, x+9, color="blue", lw=3, ls='-', marker='+')  # blue line with plus markers
ax.plot(x, x+10, color="blue", lw=3, ls='-', marker='o')  # blue line with circle markers
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')  # blue line with square markers
ax.plot(x, x+12, color="blue", lw=3, ls='-', marker='1')  # blue line with triangle-like marker

# marker size and color  # customizing marker appearance
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)  # small markers
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)  # medium markers
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")  # larger markers with red fill
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8,
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")  # square markers with custom fill and border
```

### Control over axis appearance

In this section we will look at controlling axis sizing properties in a matplotlib figure.

## Plot range

We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))  # creates a figure with 1 row and 3 subplots (axes) sized 12x4 inches

axes[0].plot(x, x**2, x, x**3)  # plots two curves (x^2 and x^3) on the first subplot
axes[0].set_title("default axes ranges")  # sets the title of the first subplot

axes[1].plot(x, x**2, x, x**3)  # plots the same two curves on the second subplot
axes[1].axis('tight')  # automatically tightens the axis limits to fit the data closely
axes[1].set_title("tight axes")  # sets the title of the second subplot

axes[2].plot(x, x**2, x, x**3)  # plots the same two curves on the third subplot
axes[2].set_ylim([0, 60])  # sets custom y-axis limits (0 to 60)
axes[2].set_xlim([2, 5])  # sets custom x-axis limits (2 to 5)
axes[2].set_title("custom axes range")  # sets the title of the third subplot
```
